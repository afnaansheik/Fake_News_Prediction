**Steaming**:

Steaming is the process of reducing a word to its Root word.
example: actor, actress, acting, --> act

```python
port_stem = PorterStemmer()
```

```python
def steaming(content):
  stemmed_content = re.sub('[^a-zA-Z]',' ',content) #removing numbers commas & space form content inly include text a-zA-Z
  stemmed_content = stemmed_content.lower() #convert text to lower case
  stemmed_content = stemmed_content.split() #convert text to list
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')] #port_stem will convert
  stemmed_content = ' '.join(stemmed_content)
  return stemmed_content
```

```python
news_dataset['content'] = news_dataset['content'].apply(steaming)
```

```python
print(news_dataset['content'])
```

```python
# separating the data and label
x = news_dataset['content'].values
y = news_dataset['label'].values
```

```python
print(x)
```

```python
print(y)
```

```python
# converting the textual data to numerical data
vectorize = TfidfVectorizer()
vectorize.fit(x)

x = vectorize.transform(x)
```

```python
print(x)
```

Splitting the dataset to training & testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, stratify=y, random_state=2 )
```

Training the Model: Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

Evaluation

accuracy score

```python
#  accuracy score on training data
x_train_prediction = model.predict(x_train)
# print(x_train_prediction)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# accuracy score on the test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```

###Detailed Exploratory Data Analysis (EDA)

```python
# Import libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Separate fake and real news
fake_news = news_dataset[news_dataset['label'] == 1]['text']  # Assuming 'label' column has 1 for fake news
real_news = news_dataset[news_dataset['label'] == 0]['text']  # Assuming 'label' column has 0 for real news

# Generate word clouds
def plot_word_cloud(data, title):
    wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white').generate(' '.join(data))
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title)
    plt.show()

# Plot word clouds for fake and real news
plot_word_cloud(fake_news, 'Fake News Word Cloud')
plot_word_cloud(real_news, 'Real News Word Cloud')
```

```python
from flask import Flask, request, jsonify
from flask_ngrok import run_with_ngrok
import joblib
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import numpy as np

app = Flask(__name__)
run_with_ngrok(app)  # Start ngrok when app is run

# Load model and vectorizer
model = joblib.load('/content/fake_news_model.joblib')
vectorizer = joblib.load('/content/vectorizer.joblib')

# Initialize stemmer
port_stem = PorterStemmer()

# Text preprocessing function
def preprocess_text(content):
    stemmed_content = re.sub('[^a-zA-Z]', ' ', content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    return ' '.join(stemmed_content)

# Prediction route
@app.route('/predict', methods=['POST'])
def predict():
    data = request.json['text']
    processed_text = preprocess_text(data)
    vectorized_text = vectorizer.transform([processed_text])
    prediction = model.predict(vectorized_text)
    result = "Fake" if prediction[0] == 1 else "Not Fake"
    return jsonify({"prediction": result})

if __name__ == "__main__":
    app.run()
```
